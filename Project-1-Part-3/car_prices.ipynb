```python
import pandas as pd
from pathlib import Path

car_prices_to_load=Path("car_prices.csv")
car_prices=pd.read_csv(car_prices_to_load)
car_prices.head()
```

```python
car_prices.count()
```

```python
car_prices_new=car_prices.dropna(how='any')
car_prices_new.count()
```

```python
cs_states_yearly=car_prices.set_index(["state"])["year"]
cs_states_count=car_prices.set_index(["state"])["make"]
cs_states=pd.DataFrame({"Year":cs_states_yearly,"No. of car sold":cs_states_count})
cs_states
```

```python
cp_states_1=car_prices_new.groupby(["state"])["make"].count()
cp_states_1
```

```python
cp_states_max=cp_states_1.max()
cp_states_min=cp_states_1.min()
print(cp_states_max)
print(cp_states_min)
```
